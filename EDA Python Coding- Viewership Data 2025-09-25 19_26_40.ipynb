# Introduction

This exercise focuses on analyzing viewership data collected from multiple streaming platforms. The dataset provides detailed information about customer viewing habits, including which shows are watched, the amount of time spent watching, and the platforms or devices used.

# Aim

The aim of this exercise is to perform exploratory data analysis (EDA) to uncover patterns and relationships within the dataset. By examining the data, we seek to understand customer preferences, identify trends in viewing behavior, and provide insights that can help improve user engagement across platforms.

# Variable Description for Each Column
Below is the name of the columns and their definitions:
  - DateID: Unique identifier for the date of the viewing event (typically in YYYYMMDD format).
  - CustomerID: Unique identifier for each customer in the dataset.
  - TotalTimeWatched: Total time (in minutes or seconds) the customer spent watching the video.
  - Platform: The streaming platform or device used to watch the video (e.g., mobile, web, smart TV).
  - PlayEventType: Type of viewing event (e.g., play, pause, stop, rewind).
  - VideoTitle: Title of the video content watched by the customer.

```python
import matplotlib.pyplot as plt

# Histogram of TotalTimeWatched
plt.figure(figsize=(8, 5))
df['TotalTimeWatched'].hist(bins=30)
plt.title('Distribution of TotalTimeWatched')
plt.xlabel('TotalTimeWatched')
plt.ylabel('Frequency')
display(plt.gcf())
plt.clf()

# Bar plot of Platform counts
plt.figure(figsize=(8, 5))
df['Platform'].value_counts().plot(kind='bar')
plt.title('Platform Distribution')
plt.xlabel('Platform')
plt.ylabel('Count')
display(plt.gcf())
plt.clf()

# Boxplot of TotalTimeWatched by Platform
plt.figure(figsize=(10, 6))
df.boxplot(column='TotalTimeWatched', by='Platform')
plt.title('TotalTimeWatched by Platform')
plt.suptitle('')
plt.xlabel('Platform')
plt.ylabel('TotalTimeWatched')
display(plt.gcf())
plt.clf()
```

```python
# prompt: create a pie chart to show users distributed by platform from the dataset 

import matplotlib.pyplot as plt

platform_counts = df['Platform'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(platform_counts, labels=platform_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Customer Distribution by Platform')
display(plt.gcf())
plt.clf()
```

# Insight: Platform distribution by  number of customers
 - The pie chart in "Platform distribution by number of customers" reveals that Leanback is the dominant platform, accounting for the largest share of unique customers. 
 - In contrast, iOS has the lowest customer representation, indicating a significant disparity in platform preference. 
 - This suggests that engagement strategies should prioritize Leanback, while further investigation is needed to understand the low adoption of iOS.
